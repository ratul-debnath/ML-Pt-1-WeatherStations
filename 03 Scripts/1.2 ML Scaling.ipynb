# 02 Import Dataset

```python
path = r'C:\Users\Ratul\anaconda_projects\Machine Learning'
df = pd.read_csv(os.path.join(path, '02 Data', 'Original Data', 'Dataset-weather-prediction-dataset-processed.csv'), index_col = False)
```

# 03 Scaling

```python
df.head()
```

```python
# Exclude the date column
df_no_date = df.drop(columns=['DATE'])
```

```python
#Select numeric columns for scaling
numeric_cols = df_no_date.select_dtypes(include=['float64', 'int64']).columns
```

```python
#Initialize scaler
scaler = StandardScaler()
```

```python
#Fit and transform the data
df_scaled = df.copy()
df_scaled[numeric_cols] = scaler.fit_transform(df_no_date[numeric_cols])
```
