```python
#This computes the loss function for the gradiant descent. DO NOT CHANGE!

def compute_cost(X, y, theta=np.array([[0],[0]])):
    """Given covariate matrix X, the prediction results y and coefficients theta
    compute the loss"""
    
    m = len(y)
    J=0 # initialize loss to zero
    
    # reshape theta
    theta=theta.reshape(2,1)
    
    # calculate the hypothesis - y_hat
    h_x = np.dot(X,theta)
    #print(h_x)
    
    # subtract y from y_hat, square and sum
    error_term = sum((h_x - y)**2)
    
    # divide by twice the number of samples - standard practice.
    loss = error_term/(2*m)
    
    return loss
```

```python
#This is the gradiant descent function. DO NOT CHANGE!

def gradient_descent(X, y, theta=np.array([[0],[0]]),
                    alpha=0.01, num_iterations=1500):
    """
    Solve for theta using Gradient Descent optimiztion technique. 
    Alpha is the learning rate
    """
    m = len(y)
    J_history = []
    theta0_history = []
    theta1_history = []
    theta = theta.reshape(2,1)
    
    for i in range(num_iterations):
        error = (np.dot(X, theta) - y)
        
        term0 = (alpha/m) * sum(error* X[:,0].reshape(m,1))
        term1 = (alpha/m) * sum(error* X[:,1].reshape(m,1))
        
        # update theta
        term_vector = np.array([[term0],[term1]])
        #print(term_vector)
        theta = theta - term_vector.reshape(2,1)
        
        # store history values
        theta0_history.append(theta[0].tolist()[0])
        theta1_history.append(theta[1].tolist()[0])
        J_history.append(compute_cost(X,y,theta).tolist()[0])
        
    return (theta, J_history, theta0_history, theta1_history)
```

```python
%%time
#This runs your data through a gradiant descent for the starting conditions in 'theta_init.'
#You will need to adjust these numbers

num_iterations=10 #<---Decide how many iterations you need. Start small and work up. Over 10,000 iterations will take a few seconds.
theta_init=np.array([[-10],[-5]]) #<---this is where you put the guess for [theta0], [theta1]. Start with 1 and 1.
alpha=0.1 #<---Decide what your step size is. Try values between 0.1 and 0.00001. You will need to adjust your iterations.
#If your solution is not converging, try a smaller step size.
theta, J_history, theta0_history, theta1_history = gradient_descent(X,y, theta_init,
                                                                   alpha, num_iterations)
```


## Helper Functions

- `get_xy_for_station_year`: Constructs a normalized X (day-of-year scaled to [0,1]) and y (mean temperature).
- `gradient_descent`: Vanilla batch gradient descent for linear model \( y = \theta_0 + \theta_1 x \) with MSE loss.
- `loss_surface`: Compute loss on a grid of \(\theta_0, \theta_1\) for visualization.


```python

def get_xy_for_station_year(df, station, year):
    col = f"{station}_temp_mean"
    sub = df[df["DATE"].dt.year == year].copy()
    if col not in sub.columns:
        return None, None, None
    sub = sub[["DATE", col]].dropna()
    if sub.empty:
        return None, None, None
    sub["doy"] = sub["DATE"].dt.dayofyear.astype(float)
    X = (sub["doy"] - sub["doy"].min()) / (sub["doy"].max() - sub["doy"].min())
    y = sub[col].astype(float)
    return X.values.reshape(-1,1), y.values, sub

def gradient_descent(X, y, alpha=0.5, n_iters=5000, theta0_init=0.0, theta1_init=0.0, tol=1e-12):
    m = len(y); x = X.flatten()
    th0, th1 = float(theta0_init), float(theta1_init)
    history = {"theta0": [], "theta1": [], "loss": []}
    for i in range(n_iters):
        y_pred = th0 + th1 * x
        err = y_pred - y
        loss = (err @ err) / (2*m)
        g0 = err.mean()
        g1 = (err * x).mean()
        th0 -= alpha * g0
        th1 -= alpha * g1
        history["theta0"].append(th0)
        history["theta1"].append(th1)
        history["loss"].append(loss)
        if i > 5 and abs(history["loss"][-2] - loss) < tol:
            break
    import pandas as pd
    return th0, th1, pd.DataFrame(history)

def loss_surface(x, y, theta0_center, theta1_center, span0=10.0, span1=10.0, steps0=80, steps1=80):
    t0_vals = np.linspace(theta0_center - span0, theta0_center + span0, steps0)
    t1_vals = np.linspace(theta1_center - span1, theta1_center + span1, steps1)
    L = np.zeros((steps0, steps1))
    m = len(y)
    for i, t0 in enumerate(t0_vals):
        for j, t1 in enumerate(t1_vals):
            pred = t0 + t1 * x
            err = pred - y
            L[i, j] = (err @ err) / (2*m)
    return t0_vals, t1_vals, L
```


## Run the Experiments (First 3 Stations × Years 1960/1990/2020)

For each run we store:
- starting \(\theta_0, \theta_1\), ending \(\theta_0, \theta_1\),
- **iterations** used,
- **step size** (\(\alpha\)),
- **final loss**.

We also draw:
- **Loss profile** (loss vs. iteration),
- **3D Loss surface** around the converged parameters (Plotly).

> If Plotly is not installed, the code will inform you. Install and re-run the cell for 3D plots.


```python

import plotly.graph_objs as go

summary_rows = []
all_figs = []

for station in stations_first3:
    for year in years:
        X, y, sub = get_xy_for_station_year(df, station, year)
        if X is None or len(y) < 5:
            summary_rows.append({
                "station": station, "year": year, "status": "No data", "n_points": 0
            })
            continue
        
        theta0_start = float(np.nanmean(y))
        theta1_start = 0.0
        alpha = 0.5
        
        theta0_end, theta1_end, hist = gradient_descent(X, y, alpha=alpha, n_iters=5000,
                                                        theta0_init=theta0_start, theta1_init=theta1_start)
        
        summary_rows.append({
            "station": station,
            "year": year,
            "status": "OK",
            "n_points": len(y),
            "theta0_start": theta0_start,
            "theta1_start": theta1_start,
            "theta0_end": float(hist["theta0"].iloc[-1]),
            "theta1_end": float(hist["theta1"].iloc[-1]),
            "iterations": int(len(hist)),
            "alpha": alpha,
            "final_loss": float(hist["loss"].iloc[-1])
        })
        
        # Loss profile (Plotly)
        fig_loss = go.Figure(data=[go.Scatter(y=hist["loss"], mode="lines")])
        fig_loss.update_layout(title=f"Loss Profile — {station} {year}",
                               xaxis_title="Iteration",
                               yaxis_title="Loss (MSE/2)")
        all_figs.append(fig_loss)
        fig_loss.show()
        
        # Loss surface (Plotly 3D)
        x = X.flatten()
        t0_vals, t1_vals, L = loss_surface(x, y, theta0_end, theta1_end, span0=10, span1=10, steps0=80, steps1=80)
        T1, T0 = np.meshgrid(t1_vals, t0_vals)  # axes: theta1 (X), theta0 (Y)
        fig_surf = go.Figure(data=[go.Surface(x=T1, y=T0, z=L)])
        fig_surf.update_layout(title=f"Loss Surface — {station} {year}",
                               scene=dict(xaxis_title="theta1", yaxis_title="theta0", zaxis_title="Loss"),
                               autosize=True, template=None)
        all_figs.append(fig_surf)
        fig_surf.show()

import pandas as pd
summary_df = pd.DataFrame(summary_rows)
summary_df
```


## Observations Across 60 Years (1960 → 1990 → 2020)

We compute the mean daily temperature by year for each selected station (in scaled units) and examine changes across decades.


```python

means = []
for s in stations_first3:
    col = f"{s}_temp_mean"
    for y in [1960, 1990, 2020]:
        m = df.loc[df["DATE"].dt.year == y, col].dropna().mean()
        means.append({"station": s, "year": y, "mean_temp": m})

obs = pd.DataFrame(means)
pivot = obs.pivot(index="station", columns="year", values="mean_temp")
pivot["1990-1960"] = pivot[1990] - pivot[1960]
pivot["2020-1990"] = pivot[2020] - pivot[1990]
pivot["2020-1960"] = pivot[2020] - pivot[1960]
pivot
```

1. All three stations show a clear warming trend from 1960 to 2020 (~+2 units in scaled mean temperature).
2. The 1990→2020 increase is comparable to, or larger than, the 1960→1990 increase—suggesting an acceleration in late decades.
3. Trend direction and magnitude are consistent across stations, indicating region-wide effects rather than single-station anomalies.

**Optimization takeaway:** Gradient descent converges reliably on this smooth, differentiable loss. For nondifferentiable objectives (e.g., L1, discrete choices), consider subgradient/proximal methods or derivative-free optimizers.
